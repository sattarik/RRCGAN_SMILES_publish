```python
from rdkit import Chem
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import rdDepictor
from rdkit.Chem.Draw import rdMolDraw2D
from IPython.display import SVG
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# read the DFT validated data
csv_name = './output_Regular_noscreenfinal_sampled.csv'
gen_SMILES = pd.read_csv(csv_name)

smiles = gen_SMILES ['SMILES']
des = gen_SMILES ['des_cv']
dft = gen_SMILES ['DFT_cv']
```

```python
def moltosvg(mol, molSize = (300,300), kekulize = True):
    mc = Chem.Mol(mol.ToBinary())
    if kekulize:
        try:
            Chem.Kekulize(mc)
        except:
            mc = Chem.Mol(mol.ToBinary())
    if not mc.GetNumConformers():
        rdDepictor.Compute2DCoords(mc)
    drawer = rdMolDraw2D.MolDraw2DSVG(molSize[0],molSize[1])
    drawer.DrawMolecule(mc)
    drawer.FinishDrawing()
    svg = drawer.GetDrawingText()
    return svg.replace('svg:','')

for i, smile in enumerate (smiles):
  m = Chem.MolFromSmiles(smile)
  my_svg = SVG(moltosvg(m))
  with open('./svg_smiles_generated/{}_tar{}_dft{}.svg'.format(i, round(des[i], 2), round(dft[i], 2)),'wb+') as outfile:
    outfile.write(my_svg.data.encode('utf-8'))
```
